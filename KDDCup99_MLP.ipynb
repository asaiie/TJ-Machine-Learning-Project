```python
import sklearn as sk
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import preprocessing
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
from sklearn.metrics import confusion_matrix
from sklearn.neural_network import MLPClassifier
```

```python
x = pd.read_csv('cfs_subset_train.csv')
y = pd.read_csv('cfs_subset_test.csv')
x["protocol_type"] = x["protocol_type"].astype('category')
x["protocol_type"] = x["protocol_type"].cat.codes
x["service"] = x["service"].astype('category')
x["service"] = x["service"].cat.codes
x["flag"] = x["flag"].astype('category')
x["flag"] = x["flag"].cat.codes
y["protocol_type"] = y["protocol_type"].astype('category')
y["protocol_type"] = y["protocol_type"].cat.codes
y["service"] = y["service"].astype('category')
y["service"] = y["service"].cat.codes
y["flag"] = y["flag"].astype('category')
y["flag"] = y["flag"].cat.codes
x_train = x.iloc[: , :8]
y_train = x.iloc[:, 9:]
x_test = y.iloc[: , :8]
y_test = y.iloc[:, 9:]
clf = MLPClassifier(random_state=1, max_iter=100).fit(x_train, y_train)
preds = clf.predict(x_test)
print(f"Train Accuracy: {clf.score(x_train, y_train)}")
print(f"Test Accuracy: {clf.score(x_test, y_test)}")
print('Confusion Matrix: \n', confusion_matrix(y_test, preds))
```
